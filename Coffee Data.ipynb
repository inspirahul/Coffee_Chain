```python
import seaborn as sns
```

```python
plt.style.use('ggplot')
```

```python
coffee['Profit'].hist()
```

```python
plt.figure(figsize=(26,4))
sns.barplot(x='Product', y='Profit', data=coffee)
plt.title('Mean Profit from each product')
```

```python
plt.figure(figsize=(26,4))
sns.barplot(x='Product', y='Total Expenses', data=coffee)
plt.title('Mean Total Expense from each product')
```

```python
plt.figure(figsize=(26,4))
sns.countplot(x='Product',data=coffee)
plt.title('Total Counts of Products in the entire data')
```

```python
plt.figure(figsize=(26,4))
sns.countplot(x='Market',data=coffee)
plt.title('Number of Orders from different Market/Region')
```

```python
plt.figure(figsize=(26,4))
sns.barplot(x='Market', y='Profit', data=coffee)
plt.title('Profit in different Market/Region')
```

```python
plt.figure(figsize=(26,4))
sns.barplot(x='Market', y='Total Expenses', data=coffee)
plt.title('Total Expenses in different Market/Regions')
```

```python
plt.figure(figsize=(26,4))
plt.plot(coffee['Profit'].resample('M').sum())
plt.title('Monthly Profit in Coffee Chain')
```

```python
plt.figure(figsize=(26,4))
plt.plot(coffee['Total Expenses'].resample('M').sum())
plt.title('Monthly Total Expenses in Coffee Chain')
```

```python
dt = pd.DataFrame(coffee['Profit'], coffee.index)
```

```python
dt
```

```python
yearly_dt= dt.resample('A').sum()
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x=yearly_dt.index, y='Profit', data=yearly_dt)

list=['2012', '2013']
plt.xticks(np.arange(2), list)
plt.title('Profit in Year 2012 & 2013')
```
